```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
imputer = SimpleImputer(strategy='median')
```

```python
train_df = imputer.fit_transform(train_df)
```

```python
test_df = imputer.transform(test_df)
```

```python
scaler = StandardScaler()
```

```python
scaled_train_df = scaler.fit_transform(train_df)
```

```python
scaled_test_df = scaler.transform(test_df)
```

```python
from sklearn.cluster import KMeans
```

```python
temp_kmeans = KMeans(n_clusters=2, init='k-means++', random_state=42)
```

```python
temp_kmeans.fit(scaled_train_df)
```

```python
temp_kmeans.cluster_centers_
```

```python
temp_kmeans.labels_
```

```python
temp_kmeans.inertia_
```

```python
wcss = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(scaled_train_df)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(2, 11), wcss)
plt.title('WCSS vs Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_coeff = []

for i in range(2, 11):
    final_kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    final_kmeans.fit(scaled_train_df)
    score = silhouette_score(scaled_train_df, final_kmeans.labels_)
    silhouette_coeff.append(score)
```

```python
silhouette_coeff
```

```python
plt.plot(range(2, 11), silhouette_coeff)
plt.xticks(range(2, 11))
plt.title('silhouette vs Number of Clusters')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Coefficient')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering, DBSCAN
```

```python
models = {
    'KMeansClustering' : KMeans(),
    'HierarchicalClustering' : AgglomerativeClustering(),
    'DBSCANClustering' : DBSCAN()
}
```

```python
all_wcss = []

for i in range(len(list(models))):
    temp_one_wcss_score = []
    for k in range(2, 11):
        model = list(models.values())[i]
        model.fit(scaled_train_df)
        temp_one_wcss_score.append(model.inertia_)
    all_wcss.append(temp_one_wcss_score)
```

```python
all_silhoutte_coeff = []

for i in range(len(list(models))):
    temp_one_score = []
    for k in range(2, 11):
        model = list(models.values())[i]
        model.fit(scaled_train_df)
        sscore = silhouette_score(scaled_train_df, model.labels_)
        temp_one_score.append(sscore)
    all_silhoutte_coeff.append(temp_one_score)
```

```python
all_silhoutte_coeff
```

```python
plt.plot(range(2, 11), all_silhoutte_coeff[0])
plt.xticks(range(2, 11))
plt.title('KMeans silhouette vs Number of Clusters')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Coefficient')
plt.show()
```

```python
plt.plot(range(2, 11), all_silhoutte_coeff[1])
plt.xticks(range(2, 11))
plt.title('Hierarchical silhouette vs Number of Clusters')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Coefficient')
plt.show()
```

```python
plt.plot(range(2, 11), all_silhoutte_coeff[2])
plt.xticks(range(2, 11))
plt.title('DBScan silhouette vs Number of Clusters')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Coefficient')
plt.show()
```

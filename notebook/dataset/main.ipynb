```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
imp = SimpleImputer(strategy='median')
```

```python
scaled_train_df_without_name_target = imp.fit_transform(train_df_without_name_target)
```

```python
scaled_test_df_without_name_target = imp.transform(test_df_without_name)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2, init='k-means++', random_state=42)
```

```python
y_train = kmeans.fit(scaled_train_df_without_name_target)
```

```python
kmeans.cluster_centers_
```

```python
y_labels = kmeans.labels_
```

```python
kmeans.inertia_
```

```python
wcss = []

for k in range(1, 21):
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(scaled_train_df_without_name_target)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1, 21), wcss)
plt.title('WCSS vs Number of Clusters')
plt.xticks(range(1, 21))
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
plt.scatter(scaled_train_df_without_name_target[:, 0], scaled_train_df_without_name_target[:, 1], c=y_labels)
```

```python
final_kmean = KMeans(n_clusters=2, init='k-means++', random_state=42)
```

```python
final_ans = final_kmean.fit_predict(scaled_test_df_without_name_target)
```

```python
from kneed import KneeLocator
```

```python
final_k = KneeLocator(range(1, 21), wcss, curve='convex', direction='decreasing')
```

```python
final_k.elbow
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_coef = []

for i in range(2, 21):
    last_kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    last_kmeans.fit(scaled_train_df_without_name_target)
    silhouette_avg = silhouette_score(scaled_train_df_without_name_target, last_kmeans.labels_)
    silhouette_coef.append(silhouette_avg)
```

```python
silhouette_coef
```

```python
plt.plot(range(2, 21), silhouette_coef)
plt.xticks(range(2, 21))
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Coefficient')
plt.show()
```
